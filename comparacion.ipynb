```python
import h5py
import pylab as pl

# Leemos el primer archivo
f1 = h5py.File('electromiograma/brazo/flecha/MOVER BRAZO FLECHA.h5')["84:BA:20:5E:FF:67"]["raw"]["channel_1"][:, 0]

# Leemos el segundo archivo
f2 = h5py.File('electromiograma/brazo/german/MOVER BRAZO GERMAN.h5')["84:BA:20:5E:FF:67"]["raw"]["channel_1"][:, 0]

f3=h5py.File('electromiograma/brazo/CARLOS A1/OpenSignals (r)evolution Report_2024-03-04_17-19-53.pdf.h5')["84:BA:20:5E:FF:67"]["raw"]["channel_1"]

f4=h5py.File('electromiograma/brazo/alex/MOVER BRAZO ALEX.h5')["84:BA:20:5E:FF:67"]["raw"]["channel_1"]

# Graficamos ambas seales
pl.plot(f1, label='Seal 1')
pl.plot(f2, label='Seal 2')
pl.plot(f3, label='Seal 3')
pl.plot(f4, label='Seal 4')

# Configuramos los lmites del eje x
pl.xlim(0,200)

# Configuramos los lmites del eje y
pl.ylim(200, 800)
# Mostramos la leyenda
pl.legend()

# Mostramos la grfica
pl.show()
```
