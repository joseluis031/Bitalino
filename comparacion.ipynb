```python
import h5py
import pylab as pl

# Leemos el primer archivo
f1 = h5py.File('electromiograma/brazo/flecha/MOVER BRAZO FLECHA.h5')["84:BA:20:5E:FF:67"]["raw"]["channel_1"][:, 0]

# Leemos el segundo archivo
f2 = h5py.File('electromiograma/brazo/german/MOVER BRAZO GERMAN.h5')["84:BA:20:5E:FF:67"]["raw"]["channel_1"][:, 0]

# Graficamos ambas señales
pl.plot(f1, label='Señal 1')
pl.plot(f2, label='Señal 2')

# Configuramos los límites del eje x
pl.xlim(0,200)

# Mostramos la leyenda
pl.legend()

# Mostramos la gráfica
pl.show()
```
